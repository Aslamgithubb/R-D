# word tokenize and stop word remove

```python
def preprocessing(textlist):
    word_token =[]
    tokenized_sentences = []

    for lists in textlist:
        word_token.append(word_tokenize(lists))

    
    stop = stopwords.words('english')
    for lst in word_token:
        for token in lst: 
            if token not in stop:
                tokenized_sentences.append(token)
                
    return tokenized_sentences

print(preprocessing(newlist))
```

```python
def untokenize(words):
    text = ' '.join(words)
    step1 = text.replace("`` ", '"').replace(" ''", '"')
    step2 = re.sub(r' ([.,:;?!%]+)([ \'"`])', r"\1\2", step1)
    return step2.strip()
```
